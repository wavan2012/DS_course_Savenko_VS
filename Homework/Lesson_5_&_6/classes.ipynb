```python
import time
from  abc import ABC, abstractmethod

def lead_time(function):
    def wrapper(self):
        a = time.time()
        function(self)
        b = time.time()
        print(f'lead_time: {b-a}')
    return wrapper

def sv_data(function):
    def wrapper(self):
        file = open('logs.txt', 'w')
        ab:str = function(self)
        file.write(f'<{function.__name__}> : {ab}')
        file.close()
    return wrapper

class Animal(ABC):

    def __init__(self, nickname: str, xxx: str, color: str, size: int, power: float, state='alive', satiety=50, kind=None, max_power=None, parent_1='not_inst', parent_2='not_inst'):
        self.nickname = nickname
        self.xxx = xxx
        self.color = color
        self.size = size
        self.power = power
        self.speed = lambda x: self.power ** 2 / (self.size * x)
        self.state = state
        self.satiety = satiety
        self.kind = kind
        self.max_power = max_power
        self.parent_1 = parent_1
        self.parent_2 = parent_2

    @sv_data
    def state_anm(self) -> str:
        return self.state

    @abstractmethod
    def status(self):
        print(self.nickname +': ' + self.state)
        a = str(self.power)
        b = str(self.satiety)
        print(f'power :{a} +satiety :{b}')

    @abstractmethod
    @lead_time
    def move(self):
        time.sleep(self.speed(10))
        print("move")


    def eat(self):
        if self.state != 'dead':
            self.satiety += 35
            print('eating')
            if self.satiety > 100 or self.satiety<=0:
                self.state = 'dead'
            if self.xxx == 'man':
                self.state = 'he is full'
            else: self.state = 'she is full'
        else: print('dead')

    def time(self):
        self.satiety -= 10
        if self.power <= self.max_power: self.power += 10

class Dog(Animal):
    def __init__(self, name:str, xxx: str, color: str):
        super().__init__(name, xxx, color, size=30, power=16, satiety=50, kind='Dog', max_power=40)


    def __add__(self, anm: Animal):
        if anm.kind == self.kind and anm.xxx != self.xxx:
         return Dog('not_inst', 'not_inst', 'not_inst')
        else: print('ERROR')

    def move(self):
        super().move()


    def status(self):
        super().status()


    def bite(self, anm: Animal):
        if anm.satiety<=0 and anm.power<=0: anm.state = 'dead'
        if self.status != 'dead':
           self.time()
           if anm.state != 'dead':
              anm.time()
              if anm.kind != 'Dog': anm.power -= self.power ** 2 // anm.power + self.power
              if anm.kind == 'Horse':
                  if self.satiety > 30: self.power -= 10
                  else: self.power -= 40
              if anm.kind == 'Cat':
                  if self.satiety > 30: self.power -= 20
                  else:
                     self.power -= 25
                     self.status = 'starve'
              if anm.kind == 'Dog':
                 self.power -= anm.power ** 2 // self.power + anm.power
              if  self.power <= 0: self.status = 'dead'
           else: print(anm.nickname + ' is dead')
        else: print('dead')

class Cat(Animal):
    def __init__(self, name: str, xxx: str, color: str):
        super().__init__(name, xxx, color, size=20, power=14, satiety=50,  kind='Cat', max_power=30)

    def __add__(self, anm: Animal):
        if anm.kind == self.kind and anm.xxx != self.xxx:
            return Dog('not_inst', 'not_inst', 'not_inst')
        else: print('ERROR')

    def move(self):
        super().move()


    def status(self):
        super().status()

    def slash(self, anm: Animal):
        if self.status != 'dead':
            self.time()
            if anm.state != 'dead':
                anm.time()
                if anm.kind != 'Cat': anm.power -= self.power ** 2 // anm.power
                self.status = 'alive'
                if anm.kind == 'Horse':
                    if self.satiety > 20:
                        self.power -= 10
                        anm.power -= 10
                    else: self.power -= 40
                if anm.kind == 'Cat':
                    if self.satiety > 30: self.power -= anm.power ** 2 // self.power / 2
                    else:
                        self.power -= 25
                        self.status = 'starve'
                if anm.kind == 'Dog':
                    self.power -= anm.power ** 2 // self.power / 2
                if  self.power <= 0: self.status = 'dead'
            else: print(anm.nickname + ' is dead')
        else: print('dead')

class Horse(Animal):
    def __init__(self, name:str, xxx: str, color: str):
        super().__init__(name, xxx, color, size=120, power=70, kind='Horse', max_power=100)

    def __add__(self, anm: Animal):
        if anm.kind == self.kind and anm.xxx != self.xxx:
            return Dog('not_inst', 'not_inst', 'not_inst')
        else: print('ERROR')

    def move(self):
        super().move()

    def status(self):
        super().status()


    def strike(self, anm: Animal):
        if self.status != 'dead':
            self.time()
            if anm.state != 'dead':
                anm.time()
                if anm.kind != 'Horse': anm.power -= self.power ** 2 // anm.power
                self.status = 'alive'
                if anm.kind == 'Horse':
                    if self.satiety > 30: self.power -= anm.power ** 2 // self.power / 2
                    else: self.power -= 40
                if anm.kind == 'Cat':
                    if self.satiety > 30: self.power -= 20
                    else:
                        self.power -= 25
                        self.status = 'starve'
                if anm.kind == 'Dog':
                    self.power -= anm.power ** 2 // self.power + anm.power
                if  self.power <= 0: self.status = 'dead'
            else: print(anm.nickname + ' is dead')
        else: print('dead')

class Vet():

    def status(anm: Animal):
        print(anm.nickname +': ' + anm.state)
        print('power : ' + str(anm.power) + ' satiety : ' + str(anm.satiety))
```
