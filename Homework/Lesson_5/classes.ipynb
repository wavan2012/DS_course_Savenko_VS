```python
class Animal:

    def __init__(self, name: str, xxx: str, color: str, size: str, power: float, status='alive', satiety=50, kind=None, max_power=None):
        self.nickname = name
        self.xxx = xxx
        self.color = color
        self.size = size
        self.power = power
        self.status = status
        self.satiety = satiety
        self.kind = kind
        self.max_power = max_power

    def status(self):
        print(self.nickname +': ' + self.status)
        print('power :' + str(self.power) +'satiety :' + str(self.satiety))

    def eat(self):
        if self.status != 'dead':
            self.satiety += 35
            print('eating')
            if self.satiety > 100 or self.satiety<=0:
                self.status = 'dead'
            if self.xxx == 'man':
                self.status = 'he is full'
            else: self.status = 'she is full'

        else: print('dead')

    def time(self):
        self.satiety -= 10
        if self.power <= self.max_power: self.power += 10

def status(anm: Animal):
    print(anm.nickname +': ' + anm.status)
    print('power : ' + str(anm.power) + ' satiety : ' + str(anm.satiety))

class Dog(Animal):
    def __init__(self, name:str, xxx: str, color: str):
        super().__init__(name, xxx, color, size='small', power=16, satiety=50, kind='Dog', max_power=40)

    def bite(self, anm: Animal):
        if self.status != 'dead':
           self.time()
           if anm.status != 'dead':
              anm.time()
              if anm.kind != 'Dog': anm.power -= self.power ** 2 // anm.power + self.power
              self.status = 'alive'
              if anm.kind == 'Horse':
                  if self.satiety > 30: self.power -= 10
                  else: self.power -= 40
              if anm.kind == 'Cat':
                  if self.satiety > 30: self.power -= 20
                  else:
                     self.power -= 25
                     self.status = 'starve'
              if anm.kind == 'Dog':
                 self.power -= anm.power ** 2 // self.power + anm.power
              if  self.power <= 0: self.status = 'dead'
           else: print(anm.nickname + ' is dead')
        else: print('dead')


class Cat(Animal):
    def __init__(self, name: str, xxx: str, color: str):
        super().__init__(name, xxx, color, size='small', power=14, satiety=50,  kind='Cat', max_power=30)

    def slash(self, anm: Animal):
        if self.status != 'dead':
            self.time()
            if anm.status != 'dead':
                anm.time()
                if anm.kind != 'Cat': anm.power -= self.power ** 2 // anm.power
                self.status = 'alive'
                if anm.kind == 'Horse':
                    if self.satiety > 20:
                        self.power -= 10
                        anm.power -= 10
                    else: self.power -= 40
                if anm.kind == 'Cat':
                    if self.satiety > 30: self.power -= anm.power ** 2 // self.power / 2
                    else:
                        self.power -= 25
                        self.status = 'starve'
                if anm.kind == 'Dog':
                    self.power -= anm.power ** 2 // self.power / 2
                if  self.power <= 0: self.status = 'dead'
            else: print(anm.nickname + ' is dead')
        else: print('dead')

class Horse(Animal):
    def __init__(self, name:str, xxx: str, color: str):
        super().__init__(name, xxx, color, size='big', power=70, kind='Horse', max_power=100)

    def strike(self, anm: Animal):
        if self.status != 'dead':
            self.time()
            if anm.status != 'dead':
                anm.time()
                if anm.kind != 'Horse': anm.power -= self.power ** 2 // anm.power
                self.status = 'alive'
                if anm.kind == 'Horse':
                    if self.satiety > 30: self.power -= anm.power ** 2 // self.power / 2
                    else: self.power -= 40
                if anm.kind == 'Cat':
                    if self.satiety > 30: self.power -= 20
                    else:
                        self.power -= 25
                        self.status = 'starve'
                if anm.kind == 'Dog':
                    self.power -= anm.power ** 2 // self.power + anm.power
                if  self.power <= 0: self.status = 'dead'
            else: print(anm.nickname + ' is dead')
        else: print('dead')

Dog_Steeve = Dog('Steve', 'man', 'beige')
Cat_Witch = Cat('Witch', 'woman', 'black')
Horse_Bayi = Horse('Bayi', 'woman', 'white')
Dog_Steeve.bite(Cat_Witch)
status(Dog_Steeve)
status(Cat_Witch)

```
